```python
import pandas as pd
import numpy as np
import math
from nsetools import Nse
nse = Nse()
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
nse = Nse()
stocks = nse.get_stock_codes()
names = []
top_50 = nse.get_preopen_nifty()
for i in range(len(top_50)):
    names.append(top_50[i]['symbol'])

print(names)
```

```python
pe = []
for i in names:
    url = requests.get(f"https://www.google.com/finance/quote/{i}:NSE")
    soup = bs4.BeautifulSoup(url.content,"html.parser")
    volume = soup.findAll(class_ = "P6K39c")
    if(volume[4].text != "-"):
        pe.append(volume[4].text)
    else:
        pe.append(None)

my_columns = ['Name','Price','Price to Earn Ratio','Number of shares to buy']
final_dataframe = pd.DataFrame(columns = my_columns)

```

```python
counts = 0
for stock in names:
        Data = nse.get_quote(stock)
        final_dataframe = final_dataframe.append(
        pd.Series(
        [
            Data['symbol'],
            Data['lastPrice'],
            float(pe[counts]) if pe[counts]!=None else None,
            'N/A',

        ],
            index = my_columns
        ),
            ignore_index = True
        )
        counts += 1
```

```python
final_dataframe.sort_values('Price to Earn Ratio',ascending=False,inplace=True)
final_dataframe
```

```python

def Calculating():
    global amount
    amount = input("Enter the total amount you want to purchase: ")
    try:
        float(amount)
    except ValueError:
        print("That's not a number!\n Please try again: ")
        amount = input("Enter the amount: ")
        float(amount)
Calculating()
size = float(amount)/len(final_dataframe.index)
```

```python
for i in range(0,len(final_dataframe.index)):
    final_dataframe.loc[i,'Number of shares to buy'] = math.floor(size/final_dataframe.loc[i,'Price'])
```
